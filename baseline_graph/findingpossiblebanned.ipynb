```python
import pandas as pd
import numpy as np
import csv
import webbrowser
```

```python
swearWords=pd.read_csv('swearWords.csv').T
print(swearWords.shape)
df = pd.DataFrame(swearWords).reset_index()
df.columns = ['idx','words']
swearWords.head()
```

```python
allSubreddits=pd.read_csv('title-data/snap-subreddit-ids-title.csv',
                  names=["subreddit", "id"])
allSubreddits.head()
```

```python
setOfPoss = set()
for index, row in allSubreddits.iterrows():
    name = row['subreddit']
    for index1, word in swearWords.iterrows():
        if word[0] in name:
            print(word[0])
            print(name)
            setOfPoss.add(name)
            break
```

```python
print(len(setOfPoss))
```

```python
with open('possiblyBanned.csv', 'w', newline='') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    wr.writerow(setOfPoss)
print('complete')
```
